```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
train2 = train.copy
test = pd.read_csv("test.csv")
```

```python
datasets = [train, test]

def mapAge(age):
    if age <= 16:
        return 0
    elif age > 16 and age <= 32:
        return 1
    elif age > 32 and age <= 48:
        return 2
    elif age > 48 and age <= 64:
        return 3
    elif age > 64:
        return 4
```

```python
for dataset in datasets: 
    dataset['Title'] = dataset['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
    
    # Changing from old names to new names
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    
    # Mapping title columns to int values
    title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4}
    dataset['Title'] = dataset['Title'].apply(lambda x: title_mapping[x] if x in title_mapping else 5)
    
    # Mapping Embarked
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
    
    # Mapping Sex columns
    dataset['Sex'] = dataset['Sex'].map( {"female": 0, "male": 1} ).astype(int)
    
    # Mapping Age from 0 to 4
    dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean()) # First replacing NaN with meanvalue
    dataset['Age'] = dataset['Age'].apply(lambda x: mapAge(x))
    dataset['Age'] = dataset['Age'].astype(int)
    
    
    filter = (dataset['SibSp']>0) & (dataset['Parch']>0)
  
    # filtering data 
    dataset['FamilySize'] = np.where((dataset['SibSp']>0) | (dataset['Parch']>0), (dataset['SibSp'] + dataset['Parch']), 0)
    dataset['IsAlone'] = np.where(dataset['FamilySize'] == 0, 1,0)
    dataset.drop(['Ticket','Fare','Cabin','SibSp', 'Parch','Name'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
y = train[['Survived']]
x = train.drop(columns=['Survived'])
```

```python
# Splitting data for training and validation
train_x, test_x, train_y, test_y = train_test_split(x, np.ravel(y))
```

```python
logreg_model = LogisticRegression()
logreg_model.fit(train_x, train_y)
preds = logreg_model.predict(test_x)
acc = accuracy_score(preds, test_y)
print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = pd.DataFrame(confusion_matrix(test_y, preds))

print(confusion_matrix)
```

```python
confusion_matrix.index = ['Actual Died','Actual Survived']
confusion_matrix.columns = ['Predicted Died','Predicted Survived']
print(confusion_matrix)
```
